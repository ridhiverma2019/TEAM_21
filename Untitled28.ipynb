```python
%%writefile pre_processing.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest

# Part 1: UCI Wine Dataset Preprocessing
#------------------------------------------

def preprocess_wine_data(wine_path):
    # Load wine dataset
    print("Loading Wine dataset...")
    # Wine data doesn't have headers in the file, so we'll add them
    column_names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
                    'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                    'Proanthocyanins', 'Color intensity', 'Hue',
                    'OD280/OD315 of diluted wines', 'Proline']

    wine_df = pd.read_csv(wine_path, names=column_names)

    print(f"Wine dataset shape: {wine_df.shape}")
    print("First few rows of the wine dataset:")
    print(wine_df.head())

    # Separate features and target
    X_wine = wine_df.drop('Class', axis=1)
    y_wine = wine_df['Class']

    # 1. Outlier Removal using Isolation Forest
    print("\nPerforming outlier removal...")
    iso_forest = IsolationForest(contamination=0.05, random_state=42)
    outlier_mask = iso_forest.fit_predict(X_wine) == 1  # 1 for inliers, -1 for outliers

    X_wine_clean = X_wine[outlier_mask]
    y_wine_clean = y_wine[outlier_mask]

    print(f"Removed {X_wine.shape[0] - X_wine_clean.shape[0]} outliers")
    print(f"Clean wine dataset shape: {X_wine_clean.shape}")

    # 2. Feature Scaling
    print("\nPerforming feature scaling...")
    scaler = StandardScaler()
    X_wine_scaled = scaler.fit_transform(X_wine_clean)

    # 3. Dimensionality Reduction with PCA
    print("\nPerforming PCA dimensionality reduction...")
    # Let's keep enough components to explain 95% of variance
    pca = PCA(n_components=0.95, random_state=42)
    X_wine_pca = pca.fit_transform(X_wine_scaled)

    print(f"Reduced dimensions: {X_wine_pca.shape[1]} components explain 95% of variance")
    print(f"Explained variance ratio: {pca.explained_variance_ratio_}")

    # Visualize data after preprocessing if number of components is 2 or 3
    if X_wine_pca.shape[1] == 2:
        plt.figure(figsize=(10, 6))
        for i in np.unique(y_wine_clean):
            plt.scatter(X_wine_pca[y_wine_clean == i, 0], X_wine_pca[y_wine_clean == i, 1],
                       label=f'Class {i}')
        plt.legend()
        plt.title('PCA of Wine Dataset (2 components)')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.show()

    return X_wine_pca, y_wine_clean, pca, scaler
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.model_selection import train_test_split
import random
from collections import deque
import skopt
from skopt import gp_minimize
from sklearn.metrics import adjusted_rand_score
from skopt.space import Real
from skopt.utils import use_named_args


np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)


class DQNAgent:
    def __init__(self, state_size, action_space, memory_size=2000, gamma=0.95,
                 epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995, learning_rate=0.001):
        self.state_size = state_size
        self.action_space = action_space
        self.action_size = len(action_space)
        self.memory = deque(maxlen=memory_size)
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.learning_rate = learning_rate
        self.model = self._build_model()
        self.target_model = self._build_model()
        self.update_target_model()

    def _build_model(self):

        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def update_target_model(self):

        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action_idx, reward, next_state, done):
        self.memory.append((state, action_idx, reward, next_state, done))

    def act(self, state, explore=True):
        if explore and np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        act_values = self.model.predict(state, verbose=0)
        return np.argmax(act_values[0])

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return
        minibatch = random.sample(self.memory, batch_size)
        states = np.array([i[0] for i in minibatch]).reshape(-1, self.state_size)
        next_states = np.array([i[3] for i in minibatch]).reshape(-1, self.state_size)


        q_values = self.model.predict(states, verbose=0)
        q_next = self.target_model.predict(next_states, verbose=0)


        for i, (state, action_idx, reward, next_state, done) in enumerate(minibatch):
            q_update = reward
            if not done:
                q_update += self.gamma * np.max(q_next[i])
            q_values[i][action_idx] = q_update


        self.model.fit(states, q_values, epochs=1, verbose=0)


        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        self.model.load_weights(name)

    def save(self, name):
        self.model.save_weights(name)



class AdaptiveHybridClustering:
    def __init__(self, X_train, n_clusters_range=(2, 10), covariance_types=None,
                 reg_covar_range=(1e-6, 1e-2), max_episodes=100):
        self.X_train = X_train


        if covariance_types is None:
            covariance_types = ['full', 'tied', 'diag', 'spherical']

        self.n_clusters_range = list(range(n_clusters_range[0], n_clusters_range[1] + 1))
        self.covariance_types = covariance_types
        self.reg_covar_values = np.logspace(np.log10(reg_covar_range[0]), np.log10(reg_covar_range[1]), 5)


        self.action_space = []
        for n_clusters in self.n_clusters_range:
            for covar_type in self.covariance_types:
                for reg_covar in self.reg_covar_values:
                    self.action_space.append({
                        'n_components': n_clusters,
                        'covariance_type': covar_type,
                        'reg_covar': reg_covar
                    })


        self.state_size = 6

        self.max_episodes = max_episodes
        self.best_hyperparameters = None
        self.best_model = None
        self.best_score = -np.inf

        self.agent = DQNAgent(state_size=self.state_size, action_space=self.action_space)


        self.bo_space = [
            Real(reg_covar_range[0], reg_covar_range[1], name='reg_covar', prior='log-uniform')
        ]

    def _get_covar_type_idx(self, covar_type):
        return self.covariance_types.index(covar_type)

    def _get_reg_covar_idx(self, reg_covar):
        idx = np.argmin(np.abs(self.reg_covar_values - reg_covar))
        return idx

    def _get_n_clusters_idx(self, n_clusters):
        return self.n_clusters_range.index(n_clusters)

    def _get_state(self, X, hyperparams=None):

        if hyperparams is None:
            n_clusters = random.choice(self.n_clusters_range)
            covar_type = random.choice(self.covariance_types)
            reg_covar = random.choice(self.reg_covar_values)
        else:
            n_clusters = hyperparams['n_components']
            covar_type = hyperparams['covariance_type']
            reg_covar = hyperparams['reg_covar']


        gmm = GaussianMixture(n_components=n_clusters, covariance_type=covar_type,
                             reg_covar=reg_covar, random_state=42)
        gmm.fit(X)
        labels = gmm.predict(X)


        if len(np.unique(labels)) > 1:
            silhouette = silhouette_score(X, labels)
            db_score = davies_bouldin_score(X, labels)
            ch_score = calinski_harabasz_score(X, labels)
        else:
            silhouette = -1.0
            db_score = 10.0
            ch_score = 0.0


        state = np.array([
            silhouette,
            1.0 - min(db_score / 10.0, 1.0),
            min(ch_score / 1000.0, 1.0),
            self._get_n_clusters_idx(n_clusters) / len(self.n_clusters_range),
            self._get_covar_type_idx(covar_type) / len(self.covariance_types),
            self._get_reg_covar_idx(reg_covar) / len(self.reg_covar_values)
        ])

        return state, labels, gmm.bic(X), silhouette

    def _calculate_reward(self, silhouette, bic, prev_silhouette=None, prev_bic=None):

        normalized_bic = np.exp(-bic / 1e6)

        reward = (silhouette + 1) / 2 + normalized_bic


        if prev_silhouette is not None and prev_bic is not None:
            silhouette_improvement = max(0, silhouette - prev_silhouette)
            bic_improvement = max(0, np.exp(-prev_bic / 1e6) - np.exp(-bic / 1e6))
            reward += silhouette_improvement + bic_improvement

        return reward

    def _fine_tune_with_bo(self, base_hyperparams, X):
        """Fine-tune reg_covar with Bayesian Optimization"""
        n_components = base_hyperparams['n_components']
        covariance_type = base_hyperparams['covariance_type']

        @use_named_args(self.bo_space)
        def objective(**params):
            gmm = GaussianMixture(
                n_components=n_components,
                covariance_type=covariance_type,
                reg_covar=params['reg_covar'],
                random_state=42
            )
            gmm.fit(X)
            bic = gmm.bic(X)
            return bic


        result = gp_minimize(
            objective,
            self.bo_space,
            n_calls=15,
            random_state=42
        )


        optimized_hyperparams = base_hyperparams.copy()
        optimized_hyperparams['reg_covar'] = result.x[0]
        return optimized_hyperparams

    def train(self, validation_fraction=0.3, batch_size=32):

        X_train, X_val = train_test_split(self.X_train, test_size=validation_fraction, random_state=42)


        for episode in range(self.max_episodes):

            current_state, _, _, _ = self._get_state(X_train)
            current_state = np.reshape(current_state, [1, self.state_size])

            prev_silhouette = None
            prev_bic = None


            steps = 10
            total_reward = 0

            for step in range(steps):

                action_idx = self.agent.act(current_state)
                hyperparams = self.action_space[action_idx]


                next_state, labels, bic, silhouette = self._get_state(X_train, hyperparams)
                next_state = np.reshape(next_state, [1, self.state_size])


                reward = self._calculate_reward(silhouette, bic, prev_silhouette, prev_bic)
                prev_silhouette, prev_bic = silhouette, bic


                done = (step == steps - 1)
                self.agent.remember(current_state[0], action_idx, reward, next_state[0], done)


                current_state = next_state
                total_reward += reward


                self.agent.replay(batch_size)


            if episode % 5 == 0:
                self.agent.update_target_model()


            current_state = np.reshape(self._get_state(X_val)[0], [1, self.state_size])
            action_idx = self.agent.act(current_state, explore=False)
            best_hyperparams = self.action_space[action_idx]


            fine_tuned_hyperparams = self._fine_tune_with_bo(best_hyperparams, X_val)


            gmm = GaussianMixture(
                n_components=fine_tuned_hyperparams['n_components'],
                covariance_type=fine_tuned_hyperparams['covariance_type'],
                reg_covar=fine_tuned_hyperparams['reg_covar'],
                random_state=42
            )
            gmm.fit(X_val)
            labels = gmm.predict(X_val)

            silhouette = silhouette_score(X_val, labels) if len(np.unique(labels)) > 1 else -1


            if silhouette > self.best_score:
                self.best_score = silhouette
                self.best_hyperparameters = fine_tuned_hyperparams
                self.best_model = gmm


            if episode % 5 == 0:
                print(f"Episode: {episode}/{self.max_episodes}, "
                      f"Reward: {total_reward:.2f}, "
                      f"Epsilon: {self.agent.epsilon:.2f}, "
                      f"Best Silhouette: {self.best_score:.4f}, "
                      f"Best Hyperparams: {self.best_hyperparameters}")


        if self.best_hyperparameters:
            self.best_model = GaussianMixture(
                n_components=self.best_hyperparameters['n_components'],
                covariance_type=self.best_hyperparameters['covariance_type'],
                reg_covar=self.best_hyperparameters['reg_covar'],
                random_state=42
            )
            self.best_model.fit(self.X_train)

        return self.best_model, self.best_hyperparameters, self.best_score

    def predict(self, X):
        if self.best_model is None:
            raise ValueError("Model not trained yet. Call train() first.")
        return self.best_model.predict(X)

    def predict_proba(self, X):
        if self.best_model is None:
            raise ValueError("Model not trained yet. Call train() first.")
        return self.best_model.predict_proba(X)



def evaluate_clustering(X, true_labels, pred_labels):
    """Evaluate clustering results using multiple metrics"""
    metrics = {}

    if len(np.unique(pred_labels)) > 1:
        metrics['silhouette'] = silhouette_score(X, pred_labels)
        metrics['davies_bouldin'] = davies_bouldin_score(X, pred_labels)
        metrics['calinski_harabasz'] = calinski_harabasz_score(X, pred_labels)
    else:
        metrics['silhouette'] = -1.0
        metrics['davies_bouldin'] = float('inf')
        metrics['calinski_harabasz'] = 0.0


    if true_labels is not None:

        from sklearn.metrics import confusion_matrix
        cm = confusion_matrix(true_labels, pred_labels)


        from scipy.optimize import linear_sum_assignment
        row_ind, col_ind = linear_sum_assignment(-cm)
        accuracy = cm[row_ind, col_ind].sum() / cm.sum()
        metrics['accuracy'] = accuracy

    return metrics

def visualize_clusters(X, labels, title="Clustering Results"):
    """Visualize clustering results in 2D"""

    if X.shape[1] > 2:
        from sklearn.decomposition import PCA
        pca = PCA(n_components=2)
        X_2d = pca.fit_transform(X)
    else:
        X_2d = X


    plt.figure(figsize=(10, 8))
    unique_labels = np.unique(labels)
    for i in unique_labels:
        plt.scatter(X_2d[labels == i, 0], X_2d[labels == i, 1], label=f'Cluster {i}', alpha=0.7)

    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()


def run_adaptive_clustering(X, y=None, n_clusters_range=(2, 10), max_episodes=100):
    """Run the adaptive hybrid clustering algorithm on the given dataset"""
    print("Initializing adaptive hybrid clustering model...")
    model = AdaptiveHybridClustering(
        X_train=X,
        n_clusters_range=n_clusters_range,
        max_episodes=max_episodes
    )

    print("Training the model...")
    best_model, best_hyperparams, best_score = model.train()

    print("\nTraining complete!")
    print(f"Best hyperparameters: {best_hyperparams}")
    print(f"Best silhouette score: {best_score:.4f}")


    pred_labels = model.predict(X)


    print("\nEvaluating clustering results...")
    metrics = evaluate_clustering(X, y, pred_labels)
    print("Clustering metrics:")
    for metric, value in metrics.items():
        print(f"- {metric}: {value:.4f}")
    ari = adjusted_rand_score(y, pred_labels)
    print(f"- Adjusted Rand Index (ARI): {ari:.4f}")


    print("\nVisualizing clustering results...")
    visualize_clusters(X, pred_labels, title=f"Clustering Results (n_components={best_hyperparams['n_components']})")

    return model, pred_labels, metrics


if __name__ == "__main__":

    from pre_processing import preprocess_wine_data


    wine_path = 'wine.data'


    print("Using Wine dataset...")
    X, y, _, _ = preprocess_wine_data(wine_path)
    n_clusters_range = (3, 10)


    model, pred_labels, metrics = run_adaptive_clustering(
        X=X,
        y=y,
        n_clusters_range=n_clusters_range,
        max_episodes=30
    )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.model_selection import train_test_split
import random
from collections import deque
import skopt
from skopt import gp_minimize
from sklearn.metrics import adjusted_rand_score
from skopt.space import Real
from skopt.utils import use_named_args

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

# Define the DQN Agent for RL hyperparameter selection
class DQNAgent:
    def __init__(self, state_size, action_space, memory_size=2000, gamma=0.95,
                 epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995, learning_rate=0.001):
        self.state_size = state_size
        self.action_space = action_space
        self.action_size = len(action_space)
        self.memory = deque(maxlen=memory_size)
        self.gamma = gamma  # discount rate
        self.epsilon = epsilon  # exploration rate
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.learning_rate = learning_rate
        self.model = self._build_model()
        self.target_model = self._build_model()
        self.update_target_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def update_target_model(self):
        # Copy weights from model to target_model
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action_idx, reward, next_state, done):
        self.memory.append((state, action_idx, reward, next_state, done))

    def act(self, state, explore=True):
        if explore and np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        act_values = self.model.predict(state, verbose=0)
        return np.argmax(act_values[0])  # returns action index

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return
        minibatch = random.sample(self.memory, batch_size)
        states = np.array([i[0] for i in minibatch]).reshape(-1, self.state_size)
        next_states = np.array([i[3] for i in minibatch]).reshape(-1, self.state_size)

        # Predict Q-values for current and next states
        q_values = self.model.predict(states, verbose=0)
        q_next = self.target_model.predict(next_states, verbose=0)

        # Update Q-values for actions taken
        for i, (state, action_idx, reward, next_state, done) in enumerate(minibatch):
            q_update = reward
            if not done:
                q_update += self.gamma * np.max(q_next[i])
            q_values[i][action_idx] = q_update

        # Train the model with updated Q-values
        self.model.fit(states, q_values, epochs=1, verbose=0)

        # Decay epsilon
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        self.model.load_weights(name)

    def save(self, name):
        self.model.save_weights(name)


# Class to manage GMM clustering with hyperparameter optimization
class AdaptiveHybridClustering:
    def __init__(self, X_train, n_clusters_range=(2, 10), covariance_types=None,
                 reg_covar_range=(1e-6, 1e-2), max_episodes=200):  # Increased max_episodes
        self.X_train = X_train

        # Define the action space (hyperparameter combinations)
        if covariance_types is None:
            covariance_types = ['full', 'tied', 'diag', 'spherical']

        self.n_clusters_range = list(range(n_clusters_range[0], n_clusters_range[1] + 1))
        self.covariance_types = covariance_types
        self.reg_covar_values = np.logspace(np.log10(reg_covar_range[0]), np.log10(reg_covar_range[1]), 5)

        # Create action space as a list of dictionaries with hyperparameter combinations
        self.action_space = []
        for n_clusters in self.n_clusters_range:
            for covar_type in self.covariance_types:
                for reg_covar in self.reg_covar_values:
                    self.action_space.append({
                        'n_components': n_clusters,
                        'covariance_type': covar_type,
                        'reg_covar': reg_covar
                    })

        # State features: [avg_silhouette, davies_bouldin, calinski_harabasz, prev_n_clusters, prev_covar_type_idx, prev_reg_covar_idx]
        self.state_size = 6

        self.max_episodes = max_episodes
        self.best_hyperparameters = None
        self.best_model = None
        self.best_score = -np.inf

        # Initialize DQN agent
        self.agent = DQNAgent(state_size=self.state_size, action_space=self.action_space)

        # Bayesian Optimization space
        self.bo_space = [
            Real(reg_covar_range[0], reg_covar_range[1], name='reg_covar', prior='log-uniform')
        ]

    def _get_covar_type_idx(self, covar_type):
        return self.covariance_types.index(covar_type)

    def _get_reg_covar_idx(self, reg_covar):
        idx = np.argmin(np.abs(self.reg_covar_values - reg_covar))
        return idx

    def _get_n_clusters_idx(self, n_clusters):
        return self.n_clusters_range.index(n_clusters)

    def _get_state(self, X, hyperparams=None):
        # If no hyperparams provided, use random valid values
        if hyperparams is None:
            n_clusters = random.choice(self.n_clusters_range)
            covar_type = random.choice(self.covariance_types)
            reg_covar = random.choice(self.reg_covar_values)
        else:
            n_clusters = hyperparams['n_components']
            covar_type = hyperparams['covariance_type']
            reg_covar = hyperparams['reg_covar']

        # Fit a GMM with these hyperparams
        gmm = GaussianMixture(n_components=n_clusters, covariance_type=covar_type,
                             reg_covar=reg_covar, random_state=42)
        gmm.fit(X)
        labels = gmm.predict(X)

        # Calculate clustering metrics
        if len(np.unique(labels)) > 1:  # Ensure at least 2 clusters for metrics
            silhouette = silhouette_score(X, labels)
            db_score = davies_bouldin_score(X, labels)
            ch_score = calinski_harabasz_score(X, labels)
        else:
            silhouette = -1.0
            db_score = 10.0  # High DB score is bad
            ch_score = 0.0   # Low CH score is bad

        # Create state vector with normalized values
        state = np.array([
            silhouette,  # Range: [-1, 1]
            1.0 - min(db_score / 10.0, 1.0),  # Invert and normalize DB score
            min(ch_score / 1000.0, 1.0),  # Normalize CH score
            self._get_n_clusters_idx(n_clusters) / len(self.n_clusters_range),  # Normalize cluster index
            self._get_covar_type_idx(covar_type) / len(self.covariance_types),  # Normalize covar type index
            self._get_reg_covar_idx(reg_covar) / len(self.reg_covar_values)  # Normalize reg_covar index
        ])

        return state, labels, gmm.bic(X), silhouette

    def _calculate_reward(self, silhouette, bic, prev_silhouette=None, prev_bic=None):
        # Balance between silhouette score (higher is better) and BIC (lower is better)
        # Normalize BIC to be in a similar range as silhouette
        normalized_bic = np.exp(-bic / 1e6)  # Adjust scaling factor based on your BIC magnitude

        reward = (silhouette + 1) / 2 + normalized_bic  # silhouette in [-1,1], normalize to [0,1]

        # Add improvement bonus
        if prev_silhouette is not None and prev_bic is not None:
            silhouette_improvement = max(0, silhouette - prev_silhouette)
            bic_improvement = max(0, np.exp(-prev_bic / 1e6) - np.exp(-bic / 1e6))
            reward += silhouette_improvement + bic_improvement

        return reward

    def _fine_tune_with_bo(self, base_hyperparams, X):
        """Fine-tune reg_covar with Bayesian Optimization"""
        n_components = base_hyperparams['n_components']
        covariance_type = base_hyperparams['covariance_type']

        @use_named_args(self.bo_space)
        def objective(**params):
            gmm = GaussianMixture(
                n_components=n_components,
                covariance_type=covariance_type,
                reg_covar=params['reg_covar'],
                random_state=42
            )
            gmm.fit(X)
            bic = gmm.bic(X)
            return bic  # We want to minimize BIC

        # Run Bayesian Optimization
        result = gp_minimize(
            objective,
            self.bo_space,
            n_calls=15,
            random_state=42
        )

        # Return the optimized hyperparameters
        optimized_hyperparams = base_hyperparams.copy()
        optimized_hyperparams['reg_covar'] = result.x[0]
        return optimized_hyperparams

    def train(self, validation_fraction=0.3, batch_size=32):
        # Split data into train and validation
        X_train, X_val = train_test_split(self.X_train, test_size=validation_fraction, random_state=42)

        # Training loop
        for episode in range(self.max_episodes):
            # Initialize environment with random state
            current_state, _, _, _ = self._get_state(X_train)
            current_state = np.reshape(current_state, [1, self.state_size])

            prev_silhouette = None
            prev_bic = None

            # Episode steps (assume fixed number of steps per episode)
            steps = 10
            total_reward = 0

            for step in range(steps):
                # Select action (hyperparameter combination)
                action_idx = self.agent.act(current_state)
                hyperparams = self.action_space[action_idx]

                # Apply action and get next state
                next_state, labels, bic, silhouette = self._get_state(X_train, hyperparams)
                next_state = np.reshape(next_state, [1, self.state_size])

                # Calculate reward
                reward = self._calculate_reward(silhouette, bic, prev_silhouette, prev_bic)
                prev_silhouette, prev_bic = silhouette, bic

                # Store transition in replay memory
                done = (step == steps - 1)
                self.agent.remember(current_state[0], action_idx, reward, next_state[0], done)

                # Update current state
                current_state = next_state
                total_reward += reward

                # Train the agent by replaying experiences
                self.agent.replay(batch_size)

            # After each episode, update target network
            if episode % 5 == 0:
                self.agent.update_target_model()

            # Evaluate on validation set with best hyperparameters from current policy
            current_state = np.reshape(self._get_state(X_val)[0], [1, self.state_size])
            action_idx = self.agent.act(current_state, explore=False)
            best_hyperparams = self.action_space[action_idx]

            # Fine-tune with Bayesian Optimization
            fine_tuned_hyperparams = self._fine_tune_with_bo(best_hyperparams, X_val)

            # Evaluate fine-tuned hyperparameters
            gmm = GaussianMixture(
                n_components=fine_tuned_hyperparams['n_components'],
                covariance_type=fine_tuned_hyperparams['covariance_type'],
                reg_covar=fine_tuned_hyperparams['reg_covar'],
                random_state=42
            )
            gmm.fit(X_val)
            labels = gmm.predict(X_val)

            silhouette = silhouette_score(X_val, labels) if len(np.unique(labels)) > 1 else -1

            # Update best model if improved
            if silhouette > self.best_score:
                self.best_score = silhouette
                self.best_hyperparameters = fine_tuned_hyperparams
                self.best_model = gmm

            # Print progress
            if episode % 5 == 0:
                print(f"Episode: {episode}/{self.max_episodes}, "
                      f"Reward: {total_reward:.2f}, "
                      f"Epsilon: {self.agent.epsilon:.2f}, "
                      f"Best Silhouette: {self.best_score:.4f}, "
                      f"Best Hyperparams: {self.best_hyperparameters}")

        # Final training on full dataset with best hyperparameters
        if self.best_hyperparameters:
            self.best_model = GaussianMixture(
                n_components=self.best_hyperparameters['n_components'],
                covariance_type=self.best_hyperparameters['covariance_type'],
                reg_covar=self.best_hyperparameters['reg_covar'],
                random_state=42
            )
            self.best_model.fit(self.X_train)

        return self.best_model, self.best_hyperparameters, self.best_score

    def predict(self, X):
        if self.best_model is None:
            raise ValueError("Model not trained yet. Call train() first.")
        return self.best_model.predict(X)

    def predict_proba(self, X):
        if self.best_model is None:
            raise ValueError("Model not trained yet. Call train() first.")
        return self.best_model.predict_proba(X)


# Evaluation and visualization functions
def evaluate_clustering(X, true_labels, pred_labels):
    """Evaluate clustering results using multiple metrics"""
    metrics = {}

    if len(np.unique(pred_labels)) > 1:  # Ensure we have at least 2 clusters
        metrics['silhouette'] = silhouette_score(X, pred_labels)
        metrics['davies_bouldin'] = davies_bouldin_score(X, pred_labels)
        metrics['calinski_harabasz'] = calinski_harabasz_score(X, pred_labels)
    else:
        metrics['silhouette'] = -1.0
        metrics['davies_bouldin'] = float('inf')
        metrics['calinski_harabasz'] = 0.0

    # If true labels are available, calculate clustering accuracy
    if true_labels is not None:
        # Create confusion matrix
        from sklearn.metrics import confusion_matrix
        cm = confusion_matrix(true_labels, pred_labels)

        # Calculate accuracy using Hungarian algorithm for optimal assignment
        from scipy.optimize import linear_sum_assignment
        row_ind, col_ind = linear_sum_assignment(-cm)
        accuracy = cm[row_ind, col_ind].sum() / cm.sum()
        metrics['accuracy'] = accuracy

    return metrics

def visualize_clusters(X, labels, title="Clustering Results"):
    """Visualize clustering results in 2D"""
    # If more than 2 dimensions, apply PCA
    if X.shape[1] > 2:
        from sklearn.decomposition import PCA
        pca = PCA(n_components=2)
        X_2d = pca.fit_transform(X)
    else:
        X_2d = X

    # Create scatter plot with different colors for each cluster
    plt.figure(figsize=(10, 8))
    unique_labels = np.unique(labels)
    for i in unique_labels:
        plt.scatter(X_2d[labels == i, 0], X_2d[labels == i, 1], label=f'Cluster {i}', alpha=0.7)

    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Main execution function
def run_adaptive_clustering(X, y=None, n_clusters_range=(2, 10), max_episodes=100):
    """Run the adaptive hybrid clustering algorithm on the given dataset"""
    print("Initializing adaptive hybrid clustering model...")
    model = AdaptiveHybridClustering(
        X_train=X,
        n_clusters_range=n_clusters_range,
        max_episodes=max_episodes
    )

    print("Training the model...")
    best_model, best_hyperparams, best_score = model.train()

    print("\nTraining complete!")
    print(f"Best hyperparameters: {best_hyperparams}")
    print(f"Best silhouette score: {best_score:.4f}")

    # Make predictions
    pred_labels = model.predict(X)

    # Evaluate results
    print("\nEvaluating clustering results...")
    metrics = evaluate_clustering(X, y, pred_labels)
    print("Clustering metrics:")
    for metric, value in metrics.items():
        print(f"- {metric}: {value:.4f}")
    ari = adjusted_rand_score(y, pred_labels)
    print(f"- Adjusted Rand Index (ARI): {ari:.4f}")

    # Print data points in each cluster
    for cluster in np.unique(pred_labels):
        print(f"\nData points in Cluster {cluster}:")
        cluster_points = X[pred_labels == cluster]
        print(cluster_points)

    # Visualize results
    print("\nVisualizing clustering results...")
    visualize_clusters(X, pred_labels, title=f"Clustering Results (n_components={best_hyperparams['n_components']})")

    return model, pred_labels, metrics

# Example usage
if __name__ == "__main__":
    # Import preprocessed data
    from pre_processing import preprocess_wine_data

    # Set paths to your data files
    wine_path = 'wine.data'

    # Using Wine dataset
    print("Using Wine dataset...")
    X, y, _, _ = preprocess_wine_data(wine_path)
    n_clusters_range = (3, 10)  # Wine has 3 true classes

    # Run adaptive clustering
    model, pred_labels, metrics = run_adaptive_clustering(
        X=X,
        y=y,
        n_clusters_range=n_clusters_range,
        max_episodes=30  # Reduced for faster execution
    )

    # Predefined input for prediction (adjusted to match the number of features)
    predefined_features = [13.5, 1.5, 2.5, 19.0, 100, 2.5, 1.0, 0.4, 1.5, 5.0]  # 10 features

    # Convert to numpy array and reshape for prediction
    features_array = np.array(predefined_features).reshape(1, -1)

    # Predict the cluster
    cluster = model.predict(features_array)
    print(f"The predicted cluster for the input wine sample is: {cluster[0]}")

    # Visualize the clustering results again after prediction
    visualize_clusters(X, pred_labels, title=f"Clustering Results (n_components={metrics['silhouette']})")





```
